```python
import sqlite3
import pandas as pd

# Connect to the database
db_path = "lastfm_dataset_2020.db"
conn = sqlite3.connect(db_path)

# Load tags table (one-hot encoded)
tags_df = pd.read_sql("SELECT * FROM tags;", conn)
metadata_df = pd.read_sql("SELECT * FROM metadata;", conn)

# Close connection
conn.close()

# Print sample
print(tags_df.dtypes)
print(metadata_df.dtypes)
```

```python
track_ids = tags_df["id_dataset"]
tag_features = tags_df.drop(columns=["id_dataset"])
```

```python
metadata_df.head(10)
```

```python
from sklearn.neighbors import NearestNeighbors

# Train k-NN model on tag features
knn = NearestNeighbors(n_neighbors=5, metric="cosine", algorithm="brute")
knn.fit(tag_features)
```

```python
def recommend_similar_tracks(track_id, num_recommendations=5):
    # Find the index of the track in the dataset
    if track_id not in track_ids.values:
        return "Track ID not found in dataset."

    track_index = track_ids[track_ids == track_id].index[0]
    track_vector = tag_features.iloc[track_index].values.reshape(1, -1)

    # Find the nearest neighbors
    distances, indices = knn.kneighbors(track_vector, n_neighbors=num_recommendations + 1)  # +1 to exclude itself

    # Get recommended track IDs (excluding the first one, which is itself)
    recommended_ids = track_ids.iloc[indices[0][1:]].values

    # Merge with metadata to get track names & artists
    recommendations = metadata_df[metadata_df["id_dataset"].isin(recommended_ids)][["id_dataset", "name", "artist"]]

    return recommendations

# Example: Recommend similar songs for a track ID
example_track_id = "561e55c4c72ff7b862a56557ae1fe089"
print(recommend_similar_tracks(example_track_id, 10))
```

```python
import pickle

# Save k-NN model to a file
with open("knn_model.pkl", "wb") as f:
    pickle.dump(knn, f)

# Save track IDs and tag features (needed for querying later)
metadata_df[["id_dataset", "id_spotify"]].to_csv("track_ids.csv", index=False)
tag_features.to_csv("tag_features.csv", index=False)
```
